```python
import numpy as np
import pandas as pd
import glob
```

```python
df = pd.read_csv("C:/Data Analytics/Automation/marketing_data.csv", parse_dates=["date"])

channel_summary = df.groupby("channel").agg(
    Leads=("lead_id", "count"),
    Revenue=("revenue", "sum"),
    Cost=("cost", "sum")
).reset_index()

channel_summary["Customers"] = (
    df.groupby("channel")["stage"]
      .apply(lambda x: (x == "Customer").sum())
      .values
)

channel_summary["conversion_rate_pct"] = (
    channel_summary["Customers"] / channel_summary["Leads"] * 100
).round(2)

channel_summary["CAC"] = (
    channel_summary["Cost"] / channel_summary["Customers"].replace(0, np.nan)
).round(2)

print(channel_summary.sort_values("Revenue", ascending=False))
```
